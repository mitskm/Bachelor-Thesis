```python
import re
from itertools import product, combinations
from collections import Counter

# ---------- 1. Define full design ----------

attributes = ["C", "E", "F", "S"]
levels = [1, 2]

# All 16 profiles
all_profiles = [
    "".join(f"{a}{l}" for a, l in zip(attributes, combo))
    for combo in product(levels, repeat=4)
]
all_profiles_set = set(all_profiles)

# All 120 unordered pairs
all_valid_pairs = {
    tuple(sorted(pair))
    for pair in combinations(all_profiles, 2)
}

# ---------- 2. Parsing ----------

profile_pattern = re.compile(r"^C[12]E[12]F[12]S[12]$")

def parse_pair(pair_string):
    left, right = pair_string.split("/")
    return left, right

def canonical_pair(left, right):
    return tuple(sorted((left, right)))

# ---------- 3. Main check ----------

observed_pairs = []
invalid_rows = []
self_pairs = []

profile_counts = Counter()
attribute_counts = Counter()

for i, row in enumerate(randomized_pairs):
    pair_string = row[0]
    
    try:
        left, right = parse_pair(pair_string)
    except:
        invalid_rows.append((i, pair_string, "bad format"))
        continue
    
    # Validate profiles
    if not profile_pattern.match(left) or not profile_pattern.match(right):
        invalid_rows.append((i, pair_string, "invalid profile encoding"))
        continue
    
    if left not in all_profiles_set or right not in all_profiles_set:
        invalid_rows.append((i, pair_string, "profile not in design"))
        continue
    
    # Check no self-pairs
    if left == right:
        self_pairs.append((i, pair_string))
        continue
    
    # Store unordered pair
    observed_pairs.append(canonical_pair(left, right))
    
    # Count profiles
    profile_counts[left] += 1
    profile_counts[right] += 1
    
    # Count attribute levels
    for prof in (left, right):
        matches = re.findall(r"([CEFS])([12])", prof)
        for attr, lvl in matches:
            attribute_counts[(attr, lvl)] += 1

# ---------- 4. Diagnostics ----------

pair_counts = Counter(observed_pairs)

# Invalid pairs (should be empty)
invalid_pairs = [
    p for p in pair_counts
    if p not in all_valid_pairs
]

# Missing pairs from 120
missing_pairs = all_valid_pairs - set(pair_counts.keys())

# Repeated pairs
repeated_pairs = {p: c for p, c in pair_counts.items() if c > 1}

# ---------- 5. Output ----------

print("\n=== BASIC CHECKS ===")
print("Total rows:", len(randomized_pairs))
print("Valid parsed pairs:", len(observed_pairs))
print("Invalid rows:", len(invalid_rows))
print("Self-pairs:", len(self_pairs))

print("\n=== DESIGN CHECK ===")
print("Unique observed pairs:", len(pair_counts))
print("Should be <= 120")
print("Invalid pairs:", len(invalid_pairs))
print("Missing pairs:", len(missing_pairs))
print("Repeated pairs:", len(repeated_pairs))

print("\n=== PROFILE BALANCE ===")
for p in sorted(all_profiles):
    print(f"{p}: {profile_counts[p]}")

print("\n=== ATTRIBUTE BALANCE ===")
for attr in attributes:
    for lvl in levels:
        print(f"{attr}{lvl}: {attribute_counts[(attr, str(lvl))]}")
```

```python
# ---------- 3. Main check ----------

observed_pairs = []
invalid_rows = []
self_pairs = []

profile_counts = Counter()
profile_counts_left = Counter()
profile_counts_right = Counter()

attribute_counts = Counter()
attribute_counts_left = Counter()
attribute_counts_right = Counter()

for i, row in enumerate(randomized_pairs):
    pair_string = row[0]
    
    try:
        left, right = parse_pair(pair_string)
    except:
        invalid_rows.append((i, pair_string, "bad format"))
        continue
    
    if not profile_pattern.match(left) or not profile_pattern.match(right):
        invalid_rows.append((i, pair_string, "invalid profile encoding"))
        continue
    
    if left not in all_profiles_set or right not in all_profiles_set:
        invalid_rows.append((i, pair_string, "profile not in design"))
        continue
    
    if left == right:
        self_pairs.append((i, pair_string))
        continue
    
    observed_pairs.append(canonical_pair(left, right))
    
    # Profile balance overall and by position
    profile_counts[left] += 1
    profile_counts[right] += 1
    profile_counts_left[left] += 1
    profile_counts_right[right] += 1
    
    # Attribute-level balance overall and by position
    for side, prof in [("left", left), ("right", right)]:
        matches = re.findall(r"([CEFS])([12])", prof)
        for attr, lvl in matches:
            attribute_counts[(attr, lvl)] += 1
            
            if side == "left":
                attribute_counts_left[(attr, lvl)] += 1
            else:
                attribute_counts_right[(attr, lvl)] += 1
```

```python
print("\n=== PROFILE BALANCE BY POSITION ===")
print("Profile | Total | Left | Right")
for p in sorted(all_profiles):
    print(
        f"{p}: "
        f"Total={profile_counts[p]}, "
        f"Left={profile_counts_left[p]}, "
        f"Right={profile_counts_right[p]}"
    )

print("\n=== ATTRIBUTE BALANCE OVERALL ===")
for attr in attributes:
    for lvl in levels:
        lvl = str(lvl)
        print(f"{attr}{lvl}: {attribute_counts[(attr, lvl)]}")

print("\n=== ATTRIBUTE BALANCE BY POSITION ===")
print("Attribute level | Total | Left | Right")
for attr in attributes:
    for lvl in levels:
        lvl = str(lvl)
        print(
            f"{attr}{lvl}: "
            f"Total={attribute_counts[(attr, lvl)]}, "
            f"Left={attribute_counts_left[(attr, lvl)]}, "
            f"Right={attribute_counts_right[(attr, lvl)]}"
        )
```

```python
import pandas as pd

from scipy.stats import chisquare, chi2_contingency

# ---------- Attribute-level left/right table ----------

attr_side_rows = []

for attr in attributes:

    for lvl in levels:

        lvl = str(lvl)

        attr_side_rows.append({

            "attribute": attr,

            "level": lvl,

            "left": attribute_counts_left[(attr, lvl)],

            "right": attribute_counts_right[(attr, lvl)]

        })

attr_side_df = pd.DataFrame(attr_side_rows)

print(attr_side_df)
print("\n=== CHI-SQUARE TESTS: ATTRIBUTE LEVEL BY POSITION ===")

for attr in attributes:
    table = pd.DataFrame({
        "left": [
            attribute_counts_left[(attr, "1")],
            attribute_counts_left[(attr, "2")]
        ],
        "right": [
            attribute_counts_right[(attr, "1")],
            attribute_counts_right[(attr, "2")]
        ]
    }, index=[f"{attr}1", f"{attr}2"])
    
    chi2_stat, p_value, dof, expected = chi2_contingency(table)
    
    print(f"\nAttribute {attr}")
    print(table)
    print(f"chi2={chi2_stat:.3f}, df={dof}, p={p_value:.4f}")
    print("\n=== CHI-SQUARE TESTS: LEFT/RIGHT BALANCE BY PROFILE ===")

for p in sorted(all_profiles):
    observed = [profile_counts_left[p], profile_counts_right[p]]
    
    # Skip profiles with zero appearances
    if sum(observed) == 0:
        continue
    
    chi2_stat, p_value = chisquare(observed)
    
    print(
        f"{p}: left={profile_counts_left[p]}, right={profile_counts_right[p]}, "
        f"chi2={chi2_stat:.3f}, p={p_value:.4f}"
    )
```
